```python
import requests
import json
import os

url = "https://<update with your hostname>/api/2.0/genie/spaces"
headers = {
    "Authorization": "Bearer <update with your PAT>",
    "Content-Type": "application/json"
}

# Read the exported JSON string from the workspace file
notebook_path = dbutils.notebook.entry_point.getDbutils().notebook().getContext().notebookPath().get()
relative_dir = os.path.dirname(notebook_path).lstrip("/")
json_file_path = os.path.join("/Workspace", relative_dir, "genie_space_import.json")

with open(json_file_path, "r") as f:
    exported_json = f.read()

payload = {
    "warehouse_id": "<update with your warehouse_id>", # enter the SQL Serverless Warehouse that will support the Genie space
    "parent_path": "/Workspace/Shared/", # update workspace path for where you want the Genie space asset imported to
    "serialized_space": exported_json,
    "title": "AI Pioneer Customer Billing Genie Space",
    "description": "This Genie Space contains customer and billing information to provide insight into structured data for the AI Pioneer lab."
}

response = requests.post(url, headers=headers, json=payload)
print(response.text)
```
